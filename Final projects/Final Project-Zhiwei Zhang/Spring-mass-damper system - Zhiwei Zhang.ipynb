# Euler's Method in Vibration System

### Spring-mass-damper System

Let's first look at the figure below.

![Image](/smd.png)

####Figure 1. SDOF system. (From The Robotics Lab http://www.roboticslab.ca/mass-spring-damper/#more-5)

Okay, you will not be surprised, since we should be familiar with it from mechanics course. This model is what we are going to discuss in this notebook.

As we know, this is a single degree-of-freedom (SDOF) system. Because there is only one independent variable required to determine the position of the mass. (Maybe SDOF system is less complicated, but let's start from the basic, then we will talk about muti-degree-of-freedom later.) A spring and a damper are attached to a mass.
$k$,$c$ are the stiffness of spring and the damping coefficient respectively. $x(t)$ represents the position of mass with respect to time from its equilibrium. $u(t)$ is an external force applied to the mass. For simplicity, $x(t)=x$,$u(t)=u$.

Hooke's law of elasticity says the extension ($x$) of a spring is directly and linearly proportional to the load ($F_s$) applied to it. Thus we have

$$
F_s = -k x
$$


In physics, damping may be mathematically modeled as a force synchronous with the velocity of the object but opposite in direction to it. The force $F_d$ is related to the velocity $v$ by

$$
F_d = -c v = -c \dot{x}
$$


According to Newton's second law that the sum of the forces $F$ on an object is equal to the mass $m$ of that object multiplied by the acceleration $a$ (second derivative of displacement) of the object

$$
\Sigma{F} = m a = m \ddot{x}
$$




We also know that the mass is affected by three forces: the spring force $F_s$, the damping force $F_d$ and the external force $u$. Therefore, the equation above can be written as

$$
F_d + F_s + u = m \ddot{x}
$$
or 
$$
m \ddot{x} + c \dot{x} + k x = u
$$


This equation is a 2nd-order differential equation. In order to solve it, two initial conditions (initial position $x_0$ and initial velocity $\dot{x_0}$) are needed. Then we will use Euler's method introduced in the phugoid model to make this ODE numerically solvable.

Now we are going to do some substitutions for the 2nd-order ODE. Let's define $x_1 = \dot{x}$, and $x_2 = \ddot{x}$. Thus $\dot{x_1} = \ddot{x} = x_2$ Then the ODE above can be expressed as

$$x_2  = - \frac{c}{m} x_1 - \frac{k}{m} x + \frac{1}{m} u$$

This will be used to calculate desired states (mass position and velocity) over each time step.
We know from Euler's method that (superscript $n$ indicates the $n$-$th$ value in the time iterations)

$$x_1^n = {\dot{x}}^n = \frac{x^{n+1}-x^n}{dt}$$


$$x_2^n = \dot{x_1^n} = \frac{x_1^{n+1}-x_1^n}{dt} =  - \frac{c}{m} x_1^n - \frac{k}{m} x^n + \frac{1}{m} u^n$$

Alright, we can arrange the equations above in a time marching way. So at each time iteration $t_n$, we can evaluate all the known data of our system to obtain the state at the next time step ($t_{n+1}$).

The equation of motion discretized with Euler's method is

$$x_1^{n+1} = (- \frac{c}{m} x_1^n - \frac{k}{m} x^n + \frac{1}{m} u^n)dt + x_1^n$$

You can see that the original 2nd-order ODE becomes 1st-order, which is called state-space method. One more thing, let's define the initial conditions so that the last thing is prepared for the coding part:
$$x_0 = 1 {m}$$ 
$$\dot{x_0} = 1 {m/s}$$ 
And the following parameters:

Mass 
$$m = 50 kg$$
Spring stiffness 
$$k = 5000 N/m$$
Damping coefficient 
$$c = 100 Ns/m$$
External sinusoidal force
$$u(t) = 500 cos(5 t)$$
The total time is $10s$, and time step $dt = 0.001s$

Now we have all resources prepared, then let's do some coding!

```python
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
from math import cos

m = 50.0               # mass
c = 100.0              # damping coefficient
k = 5000.0             # spring stiffness
T = 10.0               # final time
dt =0.001              # time increment
N = int(T/dt)+1        # number of time-steps
x0 = 1.0               # initial position
x10 = 1.0              # initial velocity
```

These form initial conditions and parameter which are used to calculate the state of mass in next time step. Let's complete the code.

```python
u = numpy.empty(N)
u[0] = 0

state = numpy.empty((N,2))
x = state[:,0]
x1 = state[:,1]
state[0] = numpy.array([x0, x10])     # initial condition

for n in range(N-1):
    u[n+1] = 500*cos(5*dt*(n+1))      # external sinusoidal force
    state[n+1,:] = numpy.array([x1[n]*dt+x[n],
                               (-c/m*x1[n]-k/m*x[n]+1/m*u[n])*dt+x1[n]])
t = numpy.linspace(0.0, T, N)         # time array

plt.figure(figsize=(16,6))
plt.grid(True)
plt.xlabel('Time (sec)')
plt.ylabel('States[$x$, $\dot{x}$]')
plt.title('Spring-mass-damper System')
plt.plot(t,state);
plt.legend(('$x$ (m)', '$\dot{x}$ (m/sec)'))
```

Finished! Here, we plot the states of the mass with respect to time. From the figure above, we can see that the curves are smooth enough due to a small time step. Both the position and velocity experience a transient period, after 3 or 4 seconds, they become relatively stable.

Now we have already learned how to solve the SDOF spring-mass-damper system by using Euler's method. Although in vibration course we can easily obtain the analytical solution for SDOF system, it is still useful to discretize the system in a numerical way.

### Compare with exact solution

The equation for spring-mass-damper system is a 2nd-order ODE and it has an exact solution of the following form:

$$
x(t) = A \frac{\omega_n}{\sqrt{1-\zeta^2}} e^{-\zeta \omega_n t} sin(\omega_n \sqrt{1-\zeta^2} t) - B \frac{1}{\sqrt{1-\zeta^2}} e^{-\zeta \omega_n t} sin(\omega_n \sqrt{1-\zeta^2} t-\phi) + C cos(\omega t) + D sin(\omega t)
$$

where $A, B, C, D, \phi$ are constants. $\zeta, \omega, \omega_n$ are damping ratio and natural frequencies respectively.

The exact solution (position function $x(t)$) is obtained by using Laplace transform. Then we apply inverse Laplace transform to convert the equation back to time domain and end up with the mass position function with respect to time. We can tell from $x(t)$ that the first two terms will exponentially decay to zero when time goes to infinity and the last two sinusoidal terms remain (as shown in the result figure).

Now we can plot our exact solution. Let's plot both the numerical solution and the exact solution to see how well Euler's method approximated the spring-mass-damper oscillations.

```python
A = 607./(229.*100.)
B = 199./229.
C = 30./229.
D = 20./(229.*5.)
zeta = 0.1
omega = 5.0
omegan = 10.0
phi = numpy.math.atan((1.0-zeta**2)**0.5/zeta)
state_exact = numpy.empty((N,1))

for n in range(N):
    state_exact[n] = A*omegan/((1.0-zeta**2)**0.5)*numpy.math.exp(-zeta*omegan*dt*n)*numpy.math.sin(omegan*((1.0-zeta**2)**0.5)*dt*n)-\
                     B/((1.0-zeta**2)**0.5)*numpy.math.exp(-zeta*omegan*dt*n)*numpy.math.sin(omegan*((1.0-zeta**2)**0.5)*dt*n-phi)+\
                     C*numpy.math.cos(omega*dt*n) + D*numpy.math.sin(omega*dt*n)
            
plt.figure(figsize=(16,6))
plt.grid(True)
plt.xlabel('Time (sec)')
plt.ylabel('Position $x(m)$')
plt.title('Spring-mass-damper System')
plt.plot(t,state[:,0])
plt.plot(t,state_exact)
plt.legend(['Numerical Solution','Analytical Solution'])
```

That shows a pretty good agreement between numerical solution and analytical solution, which indicates that Euler's method can be successfully used to calculate 2nd-order ODE for the SDOF spring-mass-damper system.

### Aside

Besides using Euler's method to numerically solve 2nd-order ODE, we can also use odeint function in scipy library. The odeint solver is intended for initial condition excitation and for external excitation where the forcing function is a continuous analytical function. We just need to change a little bit from previous code and define the equation of motion in state-space form.

```python
from scipy.integrate import odeint

def SMD(state,t):
    x = state[0]
    x1 = state[1]
    x2 = - c/m*x1 - k/m*x + 1/m*500*cos(5*t)
    return [x1, x2]

state = odeint(SMD, [x0, x10], t)

plt.figure(figsize=(16,6))
plt.grid(True)
plt.xlabel('Time (sec)')
plt.ylabel('States[$x$, $\dot{x}$]')
plt.title('Spring-mass-damper System')
plt.plot(t,state);
plt.legend(('$x$ (m)', '$\dot{x}$ (m/sec)'))
```

We can see that the two figures by using different method are almost the same.

The content below, which describes the odeint function, is from online source

(http://nbviewer.ipython.org/gist/dpsanders/d417c1ffbb76f13f678c):

"The function odeint, which is found in the integrate submodule of the scipy package, integrates ordinary differential equations of the form

$$\frac{dy}{dt} = f(y,t)$$

where $y$ is a vector in $ℝ^n$

It provides a numerical solution of the initial value problem given by this differential equation, together with an initial condition

$$y(t=t_0)=y_0$$

returning the solution curve (or trajectory) $y(t)$ as a function of time $t$, within a certain range of values of $t$.

Writing $y=(y_1,y_2,…,y_n)$, we have a system of the form

$$\dot{y_1} = f_1(y_1,y_2,…,y_n,t)$$
$$\dot{y_2} = f_2(y_1,y_2,…,y_n,t)$$
$$...$$
$$\dot{y_n} = f_n(y_1,y_2,…,y_n,t)$$

Note that in fact, any ordinary system of ODE(s) can be written in this form by defining new variables to represent higher-order derivatives,
and incorporating them into the vector $y$ of variables like we did previously. To use odeint, we must define at least the Python function $f$ which implements the mathematical function $f$. It must take a vector $y$ and a time $t$, and return the new vector $f(y,t)$."
